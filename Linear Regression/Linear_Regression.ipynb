```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(7,3), 'figure.dpi':120}) #Cài đặt kích thước cho toàn bộ các hình trong file giống nhau
```

```python
df_train = pd.read_csv(r'C:\Users\ITS\Desktop\ML\Linear Regression\train.csv')
df_train.head()
```

```python
df_train.isna().sum()
```

```python
# Vì chỉ có 1 điểm dữ liệu bị nun, nếu xóa đi cũng không ảnh hưởng tới dữ liệu nên ta bỏ luôn
df_train = df_train.dropna()
df_train.isna().sum()
```

```python
# Kiểm tra xem dữ liệu có bị duplicate hay không
df_train.duplicated().sum()
```

```python
df_train.describe()
```

```python
X = np.array(df_train['x']).reshape(-1,1) # Nếu ta không reshape lại x thì khi chạy bên dưới model.fit sẽ báo lỗi
y = df_train['y']

# Vẽ biểu đồ scatter thể hiện hai biến X , y
plt.scatter(X,y, color = 'red')
plt.xlabel('X') # Tạo label trục x
plt.ylabel('y') # Tạo label trục y
plt.grid() # Tạo đường lưới

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_ # Tính hệ số góc của model
```

```python
model.intercept_ # Tính hệ số tự do của model
```

```python
print('a=',model.coef_)
print('b=',model.intercept_)
```

**Vậy phương trình hồi quy tuyến tính cần tìm là : y = 1.00065638x - 0.10726546430095141**

```python
y_pred = model.predict(X)
y_pred
```

```python
plt.scatter(X,y, color ='red', label = 'Dữ liệu thực tế')
plt.plot(X,y_pred, color = 'black', label = 'y dự báo')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid()

plt.show()
```

```python
# Tính toán độ chính xác
from sklearn.metrics import r2_score
Accuracy = r2_score(y,y_pred)
print('The accuracy of this model is:R-square=',"{:.2f}".format(Accuracy))
```
